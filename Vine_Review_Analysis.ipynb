```python
import pandas as pd
import csv
```

```python
vine_table_df = pd.read_csv("vine_table.csv")
vine_table_df
```

```python
df1 = vine_table_df[vine_table_df["total_votes"]>19]
df1
```

```python
df2 = df1[(df1["helpful_votes"]/df1["total_votes"])>=.5]
df2
```

```python
df3 = df2[df2["vine"]=="Y"]
df3
```

```python
df4 = df2[df2["vine"]=="N"]
df4
```

```python
# total reviews
vine_table_df["review_id"].count()
```

```python
# 5 star reviews
vine_table_df["review_id"][vine_table_df["star_rating"]==5].count()
```

```python
# 5 star - Y
x = df3["review_id"][df3["star_rating"]==5].count()
y = df3["review_id"][df3["star_rating"]<5].count()
z = x/y
print(z)
```

```python
# 5 star - N
h = df4["review_id"][df4["star_rating"]==5].count()
i = df4["review_id"][df4["star_rating"]<5].count()
j = h/i
print(j)
```
